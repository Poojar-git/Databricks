```sql
CREATE TABLE employees (
    emp_id INT,
    name STRING,
    dept_id INT,
    salary INT
);
```

```sql
INSERT INTO employees VALUES
(1, 'Pooja', 101, 50000),
(2, 'Rahul', 102, 70000),
(3, 'Anita', 103, 65000),
(4, 'John', 104, 45000),
(5, 'Sara', 105, 60000);
```

```sql
CREATE TABLE departments (
    dept_id INT,
    department STRING
);
```

```sql
INSERT INTO departments VALUES
(101, 'HR'),
(102, 'IT'),
(104, 'Finance');
```

```sql
SELECT * FROM departments;
```

```sql
SELECT * FROM
employees e LEFT SEMI JOIN departments d
ON e.dept_id = d.dept_id;
```

```sql
SELECT * 
FROM employees e
WHERE EXISTS (
    SELECT 1 FROM departments d 
    WHERE e.dept_id=d.dept_id
);
```

```sql
SELECT * FROM
employees e 
LEFT ANTI JOIN
departments d
ON e.dept_id=d.dept_id;
```

```sql
SELECT * FROM 
employees e 
WHERE NOT EXISTS (
    SELECT 1 FROM departments d
    WHERE e.dept_id=d.dept_id
);
```

```sql
SELECT *
FROM employees
QUALIFY ROW_NUMBER() OVER(
    PARTITION BY dept_id
    ORDER BY salary DESC
)=1;
```

```sql
SELECT * FROM employees;
```

```sql
SELECT * FROM employees
PIVOT (
    SUM(salary)
    FOR dept_id IN (101, 102, 104, 105)
);
```
